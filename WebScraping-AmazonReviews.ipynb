```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
link="https://www.amazon.in/New-Apple-iPhone-12-128GB/product-reviews/B08L5TNJHG/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
page = requests.get(link)
```

```python
page
```

```python
page.content
```

```python
soup = bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
names=soup.find_all('span',class_='a-profile-name')
```

```python
names
```

```python
len(names)
```

```python
cust_name=[]
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
cust_name
```

```python
cust_name.pop(0)
```

```python
cust_name.pop(3)
```

```python
cust_name
```

```python
title=soup.find_all('a',class_='review-title-content')

title
```

```python
Title=[]
for i in range(0,len(title)):
    Title.append(title[i].get_text())
Title
```

```python
Title[:] = [titles.rstrip('\n') for titles in Title]
Title
```

```python
Title[:] = [titles.lstrip('\n') for titles in Title]
Title
```

```python
rating = soup.find_all('i',class_='review-rating')
rating
```

```python
rate=[]
for i in range(0,len(rating)):
    rate.append(rating[i].get_text())

rate
```

```python
len(rate)
```

```python
rate.pop(0)
```

```python
rate.pop(3)
```

```python
len(rate)
```

```python
review=soup.find_all("span",{"data-hook":"review-body"})
review
```

```python
Text= []
for i in range(0,len(review)):
    Text.append(review[i].get_text())
Text
```

```python
Text[:] = [reviews.lstrip('\n')for reviews in Text]
Text
```

```python
Text[:] = [reviews.rstrip('\n')for reviews in Text]
Text
```
